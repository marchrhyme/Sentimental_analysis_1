Clean the data

```python
import nltk  #Natural Language Toolkit (NLTK) library
import nltk.classify.util #util module within this package contains utility functions that are commonly used for classification tasks
from nltk.classify import NaiveBayesClassifier #specifically designed for building Naive Bayes classifiers
import numpy as np
import re  #regular expressions
import string
```

```python
cleanup_re = re.compile('[^a-z]+')  
#the regular expression pattern '[^a-z]+' will match one or more consecutive characters that are not lowercase letters. 
```

```python
def clean(x):
    x=str(x)  #Places input sentence to a string 
    x=x.lower() 
    x=cleanup_re.sub(' ',x).strip() 
    #substitute non-lowercase letter characters with a single space
    # strip-> Removes leading and trailing spaces from the resulting sentence
    return x

notnu['word']=notnu['reviews.text'].apply(clean)  #notnu => the not null review ratings dataset
nu['word']=nu['reviews.text'].apply(clean)     #nu for the ones containing the null review ratings
```
